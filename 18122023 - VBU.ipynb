```python
col1 = np.linspace(0, 10, 1000)
col2 = np.sin(col1)
df = pd.DataFrame({"C1" : col1, "C2" : col2})
df.head(10)
```

```python
plt.style.use('seaborn-darkgrid')
```

```python
sns.lineplot(x = df.C1, y = df.C2, data = df)
plt.show()
```

```python
plt.figure(figsize = (10, 6))
sns.set(rc = {"axes.facecolor" : "#FFFDE7", "axes.grid" : True, "xtick.labelsize" : 14, 'ytick.labelsize' : 14})
sns.lineplot(x = df.C1, y = df.C2, data = df, linewidth = 2)
sns.lineplot(x = df.C1, y = df.C2 + 1, data = df, linewidth = 2)
plt.show()
```

```python
canada = pd.read_csv("mydata/canada1.csv", delimiter = ';')
canada.head(10)
```

```python
canada.columns
```

```python
canada = canada.drop(columns = ['AREA', 'DEV', 'DevName', 'REG', 'Type', 'Coverage', 'AreaName', 'RegName'])
canada.head()
```

```python
canada.rename(columns = {'OdName' : 'Country'}, inplace = True)
canada.set_index(canada.Country, inplace = True)
canada.head()
```

```python
canada.index.name = None
canada.head()
```

```python
del canada['Country']
canada.head()
```

```python
canada = canada.transpose()
```

```python
canada.head()
```

```python
plt.figure(figsize = (16, 6))
plt.title("Immigrants from TÃ¼rkiye", fontsize = 14)
sns.set(rc = {"axes.facecolor" : '#283747', "axes.grid" : False, 'xtick.labelsize' : 10, 'ytick.labelsize' : 10})
plt.xticks(rotation = 45)
sns.lineplot(x = canada.index.values, y = canada['Turkey'])
plt.show()
```

```python
plt.figure(figsize = (16, 6))
plt.title("Immigrants from TÃ¼rkiye, China, Finland", fontsize = 14)
sns.set(rc = {"axes.facecolor" : '#283747', "axes.grid" : False, 'xtick.labelsize' : 10, 'ytick.labelsize' : 10})
plt.xticks(rotation = 45)
sns.lineplot(x = canada.index.values, y = canada['Turkey'], color = '#ff9900', label = 'TÃ¼rkiye')
sns.lineplot(x = canada.index.values, y = canada['China'], color = '#4586ff', label = 'China')
sns.lineplot(x = canada.index.values, y = canada['Finland'], color = '#a2ef44', label = 'Finland')
plt.legend(facecolor = 'grey', fontsize = 'large', edgecolor = 'black', shadow = True)
plt.show()
```

```python
insurance = pd.read_csv('mydata/insurance.csv')
insurance.head(5)
```

```python
mpl.rcParams.update(mpl.rcParamsDefault)
```

```python
plt.figure(figsize = (7,7))
sns.scatterplot(x = insurance.bmi, y = insurance.expenses)
plt.show()
```

```python
plt.figure(figsize = (7,7))
sns.scatterplot(x = insurance.bmi, y = insurance.expenses, color = '#91bd3a', marker = '>')
plt.show()
```

```python
plt.figure(figsize = (7,7))
sns.scatterplot(x = insurance.bmi, y = insurance.expenses, hue = insurance.sex)
plt.show()
```

```python
plt.figure(figsize = (7,7))
sns.scatterplot(x = insurance.bmi, y = insurance.expenses, hue = insurance.smoker)
plt.show()
```

```python
plt.figure(figsize = (7,7))
plt.xlim([0, 55])
sns.scatterplot(x = insurance.bmi, y = insurance.expenses, hue = insurance.smoker)
plt.show()
```

```python
stdperf = pd.read_csv('mydata/StudentsPerformance.csv')
stdperf.head()
```

```python
#RELPLOT
sns.set(rc = {'xtick.labelsize' : 12, 'ytick.labelsize' : 12, 'axes.labelsize' : 12, "axes.grid" : False})
sns.relplot(x = 'bmi', y = 'expenses', hue = 'sex', col = 'sex', row = 'region', data = insurance)
plt.show()
```
