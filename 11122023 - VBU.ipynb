```python
#Multiple Data
```

```python
x = np.linspace(0, 10, 2000)
plt.figure(figsize = (8,4))
plt.plot(x, np.sin(x), label = 'Sin(X)')
plt.plot(x, np.cos(x), label = 'Cos(X)')
plt.xlabel(r'$X$', fontsize = 18)
plt.ylabel(r'$Y$', fontsize = 18)
plt.title('Sin(X) & Cos(X)', fontsize = 14)
plt.legend(loc = 'upper right')
plt.show()
```

```python
x = np.linspace(0, 10, 2000)
plt.figure(figsize = (8,4))
plt.plot(x, np.sin(x), label = 'Sin(X)')
plt.plot(x, np.cos(x), label = 'Cos(X)')
plt.fill_between(x, 0, np.sin(x))
plt.fill_between(x, 0, np.cos(x))
plt.xlabel(r'$X$', fontsize = 18)
plt.ylabel(r'$Y$', fontsize = 18)
plt.title('Sin(X) & Cos(X)', fontsize = 14)
plt.legend(loc = 'lower left')
plt.show()
```

```python
plt.figure(figsize = (14,6))
x = np.linspace(0, 10, 100)
y1 = np.sin(x)
y2 = np.cos(x)
plt.subplot(1,2,1)
plt.plot(x, y1)
plt.subplot(1,2,2)
plt.plot(x, y2)
plt.show()
```

```python
plt.figure(figsize = (14,6))
x = np.linspace(0, 10, 100)
y1 = np.sin(x)
y2 = np.cos(x)
plt.subplot(2,1,1)
plt.plot(x, y1, 'b')
plt.subplot(2,1,2)
plt.plot(x, y2, 'r')
plt.show()
```

```python
#Bar Graphs (Çubuk Grafikleri)
```

```python
id1 = np.arange(1, 10)
score = np.arange(20, 110, 10)
plt.bar(id1, score)
plt.xlabel('Student ID')
plt.ylabel('Score')
plt.show()
```

```python
id1 = np.arange(1, 10)
score = np.arange(20, 110, 10)
plt.figure(figsize = (8,5))
ax = plt.axes()
ax.set_facecolor("#ECF0F1")
plt.bar(id1, score, color = '#F7FE2E')
plt.xlabel('Student ID')
plt.ylabel('Score')
plt.show()
```

```python
x1 = [1,3,5,7]
x2 = [2,4,6,8]
y1 = [7,7,7,7]
y2 = [17, 18, 29, 40]
plt.figure(figsize = (8,6))
ax = plt.axes()
ax.set_facecolor("white")
plt.bar(x1, y1, label = 'First', color = '#42B300')
plt.bar(x2, y2, label = 'Second', color = '#94E413')
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Bar Chart')
plt.legend()
plt.show()
```

```python
num1 = np.array([1,3,5,7,9])
num2 = np.array([2,4,6,8,10])
plt.figure(figsize = (8,4))
plt.bar(num1, num1**2, width = 0.2, color = '#FF6F00')
plt.bar(num2, num2**2, width = 0.2, color = '#FFB300')
plt.show()
```

```python
num1 = np.array([1,3,5,7,9])
num2 = np.array([2,4,6,8,10])
plt.figure(figsize = (8,4))
plt.bar(num1, num1**2, width = 0.2, color = '#FF6F00')
plt.bar(num2, num2**2, width = 0.2, color = '#FFB300')
for x, y in zip(num1, num1**2):
    plt.text(x, y + 0.05, '%d' % y, ha = 'center', va = 'bottom')
for x, y in zip(num2, num2**2):
    plt.text(x, y + 0.05, '%d' % y, ha = 'center', va = 'bottom')
plt.show()
```

```python
x1 = ['Doğu', 'Batı', 'Kuzey', 'Güney']
y1 = [17, 18, 29, 40]
y2 = [20, 21, 22, 23]
plt.figure(figsize = (5, 7))
plt.bar(x1, y1, label = 'Open Tickets', width = 0.5, color = '#8000FF')
plt.bar(x1, y2, label = 'Closed Tickets', width = 0.5, bottom = y1, color = '#FF6300')
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Bar Chart')
for x, y in zip(x1, y1):
    plt.text(x, y - 10, '%d' % y, ha = 'center', va = 'bottom')
    
for x, y, z in zip(x1, y2, y1):
    plt.text(x, y + z - 10, '%d' % y, ha = 'center', va = 'bottom')

plt.legend()
plt.show()
```
