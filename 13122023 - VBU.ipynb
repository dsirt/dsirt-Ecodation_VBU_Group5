```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
plt.style.use('seaborn-darkgrid')
```

```python
x = np.arange(-50, 50)
y1 = np.power(x, 2)
y2 = np.power(x, 3)
y3 = np.sin(x)
y4 = np.cos(x)
y5 = np.tan(x)
y6 = np.tanh(x)
y7 = np.sinh(x)
y8 = np.cosh(x)
y9 = np.exp(x)

fig1, ax1 = plt.subplots(nrows = 3, ncols = 3, figsize = (20, 20))
ax1[0, 0].plot(x, y1, "tab:blue")
ax1[0, 0].set_title("Square Function")
ax1[0, 0].set_xlabel('X', fontsize = 18)
ax1[0, 0].set_ylabel('Y', fontsize = 18)

ax1[0, 1].plot(x, y2, "tab:orange")
ax1[0, 1].set_title("Cubic Function")
ax1[0, 1].set_xlabel('X', fontsize = 18)
ax1[0, 1].set_ylabel('Y', fontsize = 18)

ax1[0, 2].plot(x, y3, "tab:green")
ax1[0, 2].set_title("Sine Function")
ax1[0, 2].set_xlabel('X', fontsize = 18)
ax1[0, 2].set_ylabel('Y', fontsize = 18)

ax1[1, 0].plot(x, y4, "b-")
ax1[1, 0].set_title("Cosine Function")
ax1[1, 0].set_xlabel('X', fontsize = 18)
ax1[1, 0].set_ylabel('Y', fontsize = 18)

ax1[1, 1].plot(x, y5, "r-")
ax1[1, 1].set_title("Tangent Function")
ax1[1, 1].set_xlabel('X', fontsize = 18)
ax1[1, 1].set_ylabel('Y', fontsize = 18)

ax1[1, 2].plot(x, y6, "g-")
ax1[1, 2].set_title("Hyperbolic Tangent")
ax1[1, 2].set_xlabel('X', fontsize = 18)
ax1[1, 2].set_ylabel('Y', fontsize = 18)

ax1[2, 0].plot(x, y7, "m-")
ax1[2, 0].set_title("Hyperbolic Sine")
ax1[2, 0].set_xlabel('X', fontsize = 18)
ax1[2, 0].set_ylabel('Y', fontsize = 18)

ax1[2, 1].plot(x, y8, "y-")
ax1[2, 1].set_title("Hyperbolic Cosine")
ax1[2, 1].set_xlabel('X', fontsize = 18)
ax1[2, 1].set_ylabel('Y', fontsize = 18)

ax1[2, 2].plot(x, y9, "k-")
ax1[2, 2].set_title("Exponential Function")
ax1[2, 2].set_xlabel('X', fontsize = 18)
ax1[2, 2].set_ylabel('Y', fontsize = 18)

plt.show()
```

```python
#Scatter (Saçılım) Plot 
```

```python
x1 = np.array([250, 150, 350, 252, 450, 550, 455, 358, 158, 355])
y1 = np.array([40, 50, 80, 90, 100, 50, 60, 88, 54, 45])

x2 = np.array([200, 100, 300, 220, 400, 500, 450, 380, 180, 350])
y2 = np.array([400, 500, 800, 900, 1000, 500, 600, 808, 504, 405])

#Grafik-1
plt.scatter(x1, y1)
plt.xlabel('Time Spent', fontsize = 12)
plt.ylabel('Score', fontsize = 12)
plt.title('Scatter Graph')
plt.show()
```

```python
#Grafik-2
plt.scatter(x2, y2, color = 'r')
plt.xlabel('Time Spent', fontsize = 12)
plt.ylabel('Score', fontsize = 12)
plt.title('Scatter Graph')
plt.show()
```

```python
plt.scatter(x1, y1, label = 'Class 1')
plt.scatter(x2, y2, label = 'Class 2', color = 'r')
plt.xlabel('Time Spent', fontsize = 12)
plt.ylabel('Score', fontsize = 12)
plt.title('Scatter Graph')
plt.legend()
plt.show()
```

```python
plt.scatter(x1, y1, label = 'Class 1', marker = 'o', color = 'm')
plt.scatter(x2, y2, label = 'Class 2', marker = 'v', color = 'r')
plt.xlabel('Time Spent', fontsize = 12)
plt.ylabel('Score', fontsize = 12)
plt.title('Scatter Graph')
plt.legend()
plt.show()
```

```python
plt.figure(figsize = (7,7))
area = [48, 30, 20, 15]
labels = ['Low', 'Medium', 'High', 'Critical']
colors = ['#8BC34A', '#D4E157', '#FFB300', '#FF7043']
plt.pie(area, labels = labels, colors = colors, startangle = 45, autopct = '%1.1f%%')
plt.show()
```

```python
plt.figure(figsize = (7,7))
area = [48, 30, 20, 15]
total = np.sum(area)
labels = ['Low', 'Medium', 'High', 'Critical']
colors = ['#8BC34A', '#D4E157', '#FFB300', '#FF7043']

def val_per(x):
    return '{:.2f}%\n({:.0f})'.format(x, total * x / 100)

plt.pie(area, labels = labels, colors = colors, startangle = 45, autopct = val_per, shadow = 'true')
plt.show()
```

```python
fig = plt.figure(figsize = (20,6))
area = [48, 30, 20, 15]
priority = ['Low', 'Medium', 'High', 'Critical']
status = ['Resolved', 'Cancelled', 'Pending', 'Assigned']
company = ['IBM', 'Microsoft', 'Tesla', 'Apple']
colors = ['#8BC34A', '#D4E157', '#FFB300', '#FF7043']

plt.subplot(1, 3, 1)
plt.pie(area, labels = priority, colors = colors, startangle = 45)
my_circle = plt.Circle((0,0), 0.7, color = 'white')
p = plt.gcf()
p.gca().add_artist(my_circle)

plt.subplot(1, 3, 2)
plt.pie(area, labels = status, colors = colors, startangle = 45)
my_circle = plt.Circle((0,0), 0.7, color = 'white')
p = plt.gcf()
p.gca().add_artist(my_circle)

plt.subplot(1, 3, 3)
plt.pie(area, labels = company, colors = colors, startangle = 45)
my_circle = plt.Circle((0,0), 0.7, color = 'white')
p = plt.gcf()
p.gca().add_artist(my_circle)

plt.show()
```

```python
x = np.random.normal(size = 2000)
plt.hist(x, bins = 40, color = 'yellowgreen')
plt.gca().set(title = 'Histogram', ylabel = 'Frequency')
plt.show()
```

```python
x = np.random.normal(size = 2000)
plt.hist(x, bins = 40, color = 'yellowgreen', edgecolor = '#6A9662')
plt.gca().set(title = 'Histogram', ylabel = 'Frequency')
plt.show()
```

```python
#Çoklu Histogram
plt.figure(figsize = (8, 11))
x = np.random.normal(-4, 1, size = 800)
y = np.random.normal(0, 1.5, size = 800)
z = np.random.normal(3.5, 1, size = 800)

plt.hist(x, bins = 30, color = 'yellowgreen', alpha = 0.6)
plt.hist(y, bins = 30, color = '#FF8F00', alpha = 0.6)
plt.hist(z, bins = 30, color = 'blue', alpha = 0.6)
plt.gca().set(title = 'Histogram', ylabel = 'Frequency')
plt.show()
```
